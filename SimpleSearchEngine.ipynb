```python
def lireFichier(chemin):
  fichier = open(chemin , encoding="utf8")
  contenu = fichier.read()
  fichier.close()
  return contenu
```

```python
def decoupeMots(chaine):
  return chaine.split(' ')
```

```python
import glob
index = {}
for chemin in glob.glob("document/*")[:5]:
    chaine = lireFichier(chemin)
    mots = decoupeMots(chaine)
    for mot in mots:
        if mot not in index:
            index[mot] = set()      #pour eviter les doublons 
        index[mot].add(chemin)
print(len(index)) # taille de l'index
print(list(index.keys())[:50]) # affiche tous les 50 premiere clé de notre dictionnaire 
print(index["indique"]) # Affiche tous les document ou se trouve le mot indique 
```

```python
def TrouverDocument(index,Requete):
    union=[]
    motdecoupe = Requete.split(' ')
    for mots in motdecoupe:
        if mots in index:
          print("liste des documents contenat le mots", mots )
          print(set(index[mots]))
          union=union+index[mots]
    print("ensemble des documents contenant les mots de la requete ")
    print(set(union ))
```

```python
def TrouversansDocument(index,Requete):
    union=[]
    motdecoupe = decoupeMots(Requete)
    ensembleMots= set(motdecoupe)
    print(ensembleMots)
    for mots in ensembleMots:
         if mots in index:
             print("liste des documents contenant le mots", mots )
             print(set(index[mots]))
             union=union+index[mots]
         else:
          print("le mots", mots,"ne se trouve dans aucun document")
    print("ensemble des documents contenant les mots de la requete ")
    print(set(union ))
```

```python
import re
def afficher_contextes(chaine , terme , taille_contexte = 30):
    match = re.search(terme , chaine)
    contexts = []
    while match is not None:
        #Les bornes gauche et droite du mots 
        gauche = max(match.start() - taille_contexte - 1, 0)
        droite = match.end()+1+ taille_contexte
        contexts . append(chaine[gauche:droite])
        chaine = chaine[match.end():]
        match = re. search (terme , chaine )
    for c in contexts :
        print (c)
```

```python
afficher_contextes('Commission Européenne Européenne','Commission')
```
